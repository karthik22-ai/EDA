```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
Country_names = final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=Country_names[:3],autopct='%1.2f%%')
plt.show()
```

observation: We can see that major percentage of zomato is in india the rest of other two top countries are  united states stands in the second place and third place is united kingdom. I visualized the top 3 countries which uses zomato. As zomato major base in india we can see most transactions are from india only.

```python
ratings =final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

When Rating is between 4.5 to 4.9---> Excellent  .
When Rating are between 4.0 to 3.4--->very good   .
when Rating is between 3.5 to 3.9----> good      .
when Rating is between 3.0 to 3.4----> average   .
when Rating is between 2.5 to 2.9----> average  .
when Rating is between 2.0 to 2.4----> Poor  .

```python
ratings.head()
```

```python
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings,hue = 'Rating color' , palette =['blue', 'red', 'orange', 'yellow', 'green', 'grey'])
plt.show()
```

Observation:

1)Not Rated count is very high.    
2)Maximum number of rating are between 2.5 to 3.4

```python
sns.countplot(x="Rating color",hue='Rating color', data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'grey'])
plt.show()
```
